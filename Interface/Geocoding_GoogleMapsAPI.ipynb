```python
import time
import googlemaps
import pandas as pd
```

```python
g_API = "Your-API-Key"
gmaps_key = googlemaps.Client(key=g_API)
```

```python
locations = pd.read_csv('/content/drive/MyDrive/Lok Sabha Data/Location data/LS16_location_data.csv')
```

```python
coordinates = list()
frequencies = dict()
notFound = list()
```

```python
start_time = time.time()
```

```python
for index,loc in enumerate(locations['location'].values):
  response = gmaps_key.geocode(loc)
  if len(response)>0:
    print(index, "found")
    coordinates.append([loc, response[0]['formatted_address'], response[0]['geometry']['location']['lat'], response[0]['geometry']['location']['lng']])

    if response[0]['formatted_address'] in frequencies.keys():
      frequencies[response[0]['formatted_address']] += locations.iloc[index]['frequency']
    else:
      frequencies[response[0]['formatted_address']] = locations.iloc[index]['frequency']
  else:
    notFound.append(index)
    print(index, "not found")
  print(f"Total : {index+1-len(notFound)}/{index+1}")
  print("Time elapsed :", time.time()-start_time)
  print("\n----------------------------------\n")

  if (index%1000)==0:
    plot_data = pd.DataFrame(coordinates, columns=['actual_address', 'formatted_address', 'latitude', 'longitude'])
    plot_data.drop_duplicates(subset=['formatted_address'], inplace=True, ignore_index=True)

    freqs = list()

    for loc in plot_data['formatted_address'].values:
        freqs.append(frequencies[loc])

    plot_data.insert(4, "Frequency", freqs, True)
    plot_data.to_csv('/content/drive/MyDrive/Lok Sabha Data/Location data/LS16_plot_data.csv')

    not_found_locations = list()
    for index in notFound:
        not_found_locations.append(locations.iloc[index]['location'])

    not_found = pd.DataFrame()
    not_found.insert(0, "index", notFound, True)
    not_found.insert(1, "location", not_found_locations, True)

    not_found.to_csv('/content/drive/MyDrive/Lok Sabha Data/Location data/LS16_not_found.csv')
    print(f"Saved after {index}")
```
